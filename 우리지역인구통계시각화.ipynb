# 3. 남녀 인구 추세(2019~2024)

```python
# 광진구 전체(0번행)
gwangjin_total = df.iloc[0]

# 남자 컬럼
male_cols = [c for c in df.columns if '남자 인구수' in c]
female_cols = [c for c in df.columns if '여자 인구수' in c]

# 숫자로 변환
male_values = gwangjin_total[male_cols].str.replace(',', '').astype(int).values
female_values = gwangjin_total[female_cols].str.replace(',', '').astype(int).values

# 연도 리스트
years = [int(c.split('년')[0]) for c in male_cols]

# 데이터프레임 생성
sex_df = pd.DataFrame({
    '연도': years,
    '남자': male_values,
    '여자': female_values
}).sort_values('연도')

sex_df
```

## 시각화

```python
plt.figure(figsize=(6,4))
plt.plot(sex_df['연도'], sex_df['남자'], marker='o', label='남자')
plt.plot(sex_df['연도'], sex_df['여자'], marker='o', label='여자')
plt.title("광진구 남녀 인구 추세 (2019~2024)")
plt.xlabel("연도")
plt.ylabel("인구수")
plt.legend()
plt.grid(True)
plt.show()
```

그래프 해석
- 남자 인구: 169k → 158k (약 –6.5%)
- 여자 인구: 177k → 171k (약 –3.4%)
- 감소폭: 남성 > 여성
- 꾸준한 인구 유출 + 학령인구 감소 영향 반영

행정 인사이트
- 남성 인구 감소가 더 가파르다는 점은 군입대/이주/취업 이동에 더 민감한 청년층 이동을 반영
- 여성 인구는 상대적으로 안정적 → 주거 안정 요인 추정 가능

제안 포인트
- 청년 취업·교육 프로그램 확대 (직업교육, 취업매칭 등)
- 동별 청년층 이동량이 큰 지역은 교통·생활 지원 집중
- 건대·광장동 주변에 청년 커뮤니티/창업공간 조성

# 4. 연령대별 추세

```python
house_cols = [c for c in df.columns if '세대당 인구' in c]

# 세대당 인구는 이미 float 형태라 str.replace 필요 없음!
house_values = gwangjin_total[house_cols].astype(float)

house_years = [int(c.split('년')[0]) for c in house_cols]

house_df = pd.DataFrame({
    '연도': house_years,
    '세대당_인구': house_values.values
}).sort_values('연도')

plt.figure(figsize=(6,4))
plt.plot(house_df['연도'], house_df['세대당_인구'], marker='o')
plt.title("광진구 세대당 인구 변화 (2019~2024)")
plt.xlabel("연도")
plt.ylabel("세대당 인구")
plt.grid(True)
plt.show()
```

그래프 해석
- 2019년 2.15명 → 2024년 1.96명
- 5년간 세대당 인구가 꾸준히 감소

행정 인사이트

    - 배달·편의 중심 소비 증가
    - 보건복지 필요 형태 변화(고립감, 정서지원 등)

제안 포인트
- 청년·1인가구 특화 주거정책(역세권 청년주택 연계)
- 1인가구 안심마을 조성
- 쓰레기 배출 시간 조정 등 생활SOC 미시정책 강화

# 2019~2024 연령대별 추세 변화 그래프

```python
# 분석할 연령대 리스트
age_groups = [
    "0~9세", "10~19세", "20~29세", "30~39세",
    "40~49세", "50~59세", "60~69세",
    "70~79세", "80~89세", "90~99세", "100세 이상"
]

years = range(2019, 2025)

plt.figure(figsize=(10,6))

for age in age_groups:
    vals = []
    for y in years:
        col = f"{y}년_거주자_{age}"
        val = int(str(g[col]).replace(",", ""))  # 콤마 제거 후 int 변환
        vals.append(val)
    
    plt.plot(list(years), vals, marker='o', label=age)

plt.title("광진구 연령대별 인구 변화 (2019~2024)")
plt.xlabel("연도")
plt.ylabel("인구수")
plt.grid(True)
plt.legend(title="연령대")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# df, g가 이미 정의돼 있다면 아래 두 줄은 생략해도 됨
# df = pd.read_csv("your_file.csv", encoding="cp949")
# g = df.iloc[0]   # 광진구 전체

# 4개 연령대 그룹 정의
age_groups = {
    "유소년(0~19세)": ["0~9세", "10~19세"],
    "청년(20~39세)": ["20~29세", "30~39세"],
    "중년(40~59세)": ["40~49세", "50~59세"],
    "고령(60세 이상)": ["60~69세", "70~79세", "80~89세", "90~99세", "100세 이상"],
}

years = range(2019, 2025)

plt.figure(figsize=(10,6))

for group_name, ages in age_groups.items():
    vals = []
    for y in years:
        total = 0
        for a in ages:
            col = f"{y}년_거주자_{a}"
            total += int(str(g[col]).replace(",", ""))
        vals.append(total)
    
    plt.plot(list(years), vals, marker='o', label=group_name)

plt.title("광진구 연령대 그룹별 인구 변화 (2019~2024)")
plt.xlabel("연도")
plt.ylabel("인구수")
plt.grid(True)
plt.legend(title="연령대 그룹")
plt.tight_layout()
plt.show()
```

# 청년층·고령층 비중 시계열 (2019~2024)

```python
years = list(range(2019, 2025))

# 청년층: 20~39세
young_ages = ["20~29세", "30~39세"]

# 고령층: 60세 이상
old_ages = ["60~69세", "70~79세", "80~89세", "90~99세", "100세 이상"]

young_ratio = []
old_ratio = []

for y in years:
    total = int(str(g[f"{y}년_거주자_총인구수"]).replace(",",""))
    young = sum(int(str(g[f"{y}년_거주자_{a}"]).replace(",","")) for a in young_ages)
    old = sum(int(str(g[f"{y}년_거주자_{a}"]).replace(",","")) for a in old_ages)

    young_ratio.append(young / total * 100)
    old_ratio.append(old / total * 100)

plt.figure(figsize=(8,5))
plt.plot(years, young_ratio, marker='o', label="청년층(20~39)")
plt.plot(years, old_ratio, marker='s', label="고령층(60세 이상)")
plt.title("광진구 청년층·고령층 비중 (2019~2024)")
plt.ylabel("비중 (%)")
plt.xlabel("연도")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# 동별 고령화율 랭킹 (2024년)

```python
dong = df.iloc[1:].copy()

# 전체 인구
dong['total'] = dong['2024년_거주자_총인구수'].astype(str).str.replace(",","").astype(int)

# 고령층
old_cols = [
    '2024년_거주자_60~69세', '2024년_거주자_70~79세',
    '2024년_거주자_80~89세', '2024년_거주자_90~99세',
    '2024년_거주자_100세 이상'
]

dong['old'] = dong[old_cols].apply(lambda row: 
                                   sum(int(str(x).replace(",","")) for x in row), axis=1)

dong['고령화율'] = dong['old'] / dong['total'] * 100

ranking = dong[['행정구역', '고령화율']].sort_values("고령화율", ascending=False)
ranking.head(10)
```
